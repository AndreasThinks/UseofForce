```python
taser_reg_df[logit_vars + ["Any_Taser_Use"]].corr()
```

```python
y= taser_reg_df["Any_Taser_Use"]
X = taser_reg_df[logit_vars]
```

```python
logit =  sm.Logit(y, X)
```

```python
res = logit.fit()
```

```python
res.summary()
```

```python
params = res.params
conf = res.conf_int()
conf['Odds Ratio'] = params
conf.columns = ['5%', '95%', 'Odds Ratio']
np.exp(conf).sort_values(by="Odds Ratio")
```

```python
y= taser_reg_df["Taser_Fired"]
X = taser_reg_df[logit_vars]
```

```python
logit =  sm.Logit(y, X)
```

```python
#logit =  sm.GLM(y, X, family=sm.families.Binomial())
```

```python
res.summary2()
```

```python
params = res.params
conf = res.conf_int()
conf['Odds Ratio'] = params
conf.columns = ['5%', '95%', 'Odds Ratio']
np.exp(conf)
```

## Geospatial

Let's look at where people are tasered, and where they're tasered by ethnicity

```python
taser_only = df[df["Any_Taser_Use"] == 1].copy()
```

```python
taser_black_only = df[(df["Any_Taser_Use"] == 1) & (df["SubjectEthnicity"] == "Black (or Black British)")].copy()
```

```python
taser_white_only = df[(df["Any_Taser_Use"] == 1) & (df["SubjectEthnicity"] == "White")].copy()
```

```python
borough_list = df["Borough"].unique()
```

```python
geo_data = pd.DataFrame(index=borough_list)
```

```python
geo_data["All_UoF"] = df["Borough"].value_counts()
geo_data["Taser_Use"] = taser_only["Borough"].value_counts()
geo_data["Taser_Use_black"] = taser_black_only["Borough"].value_counts()
geo_data["Taser_Use_white"] = taser_white_only["Borough"].value_counts()
geo_data["Taser_Prop"] = geo_data["Taser_Use"] / geo_data["All_UoF"]
geo_data.sort_values(by="Taser_Use")
```

```python
mopac_atlas = pd.read_excel("data/mopac_atlas.xlsx")
mopac_atlas
```

```python
#murder_dets = geomatrix[["Borough", "3_yr_murder_cnt"]].to_pickle("3_year_murder.pkl")
murder_dets = pd.read_pickle("3_year_murder.pkl")
```

```python
geomatrix = geo_data.merge(murder_dets[["Borough","3_yr_murder_cnt"]], how="left", left_index=True, right_on="Borough")
geomatrix = geomatrix.merge(mopac_atlas, how="left", left_on="Borough",right_on="Area name").dropna(axis=0)
#geomatrix.to_pickle("geomatrix.pkl")
```

```python
#geomatrix = pd.read_pickle("geomatrix.pkl")
```

```python
geomatrix = geomatrix.dropna(axis=1)
geomatrix
```
